```python
from tradingview_ta import TA_Handler, Interval, Exchange
import pandas as pd
# https://tvdb.brianthe.dev/

ticker = "EURUSD"
handler = TA_Handler(
    symbol=ticker,
    screener="forex",
    exchange="FX_IDC",
    interval=Interval.INTERVAL_1_DAY,
    # proxies={'http': 'http://example.com:8080'} # Uncomment to enable proxy (replace the URL).
)
print(f'ticker: {ticker}, {handler.get_analysis().summary}')
print(f'ticker: {ticker}, {handler.get_analysis().indicators}')
print(f'ticker: {ticker}, {handler.get_analysis().indicators["RSI"]}')
print(f'ticker: {ticker}, {handler.get_analysis().indicators["RSI[1]"]}')
```
